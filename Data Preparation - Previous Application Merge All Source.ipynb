```python
def get_grouped_data(customer_data):
    
    total_loans_applied = customer_data.shape[0]
    total_loan_approved = customer_data[customer_data["NAME_CONTRACT_STATUS_APPLICATION"]=="Approved"].shape[0]
    total_loan_rejected = customer_data[customer_data["NAME_CONTRACT_STATUS_APPLICATION"].isin(["Refused", "Canceled"])].shape[0]
    
    nbr_cash_loans = customer_data[customer_data["NAME_CONTRACT_TYPE"]=="Cash loans"].shape[0]
    nbr_consumer_loans = customer_data[customer_data["NAME_CONTRACT_TYPE"]=="Consumer loans"].shape[0]
    nbr_revolving_loans = customer_data[customer_data["NAME_CONTRACT_TYPE"]=="Revolving loans"].shape[0]
    
    nbr_live_loans = customer_data[(customer_data["NAME_CONTRACT_STATUS"]=="Active") | (customer_data["NAME_CONTRACT_STATUS_POS"]== "Active")].shape[0]
    nbr_closed_loans=  customer_data[(customer_data["NAME_CONTRACT_STATUS"] !="Active") | (customer_data["NAME_CONTRACT_STATUS_POS"] != "Active")].shape[0]
    
    nbr_loan_with_downpayment= customer_data[customer_data["RATE_DOWN_PAYMENT"]>0].shape[0]
    avg_percentage_downpayment= customer_data[customer_data["RATE_DOWN_PAYMENT"]>0]["RATE_DOWN_PAYMENT"].mean()
    
    nbr_loan_rejection_HC = customer_data[customer_data["CODE_REJECT_REASON"]=="HC"].shape[0]
    nbr_loan_rejection_LIMIT = customer_data[customer_data["CODE_REJECT_REASON"]=="LIMIT"].shape[0]
    nbr_loan_rejection_SCO = customer_data[customer_data["CODE_REJECT_REASON"].isin(["SCO", "SCOFR"])].shape[0]
    
    nbr_loan_cross_sell = customer_data[customer_data["NAME_PRODUCT_TYPE"]=="x-sell"].shape[0]
    nbr_loan_walkin = customer_data[customer_data["NAME_PRODUCT_TYPE"].isin(["walk-in"])].shape[0]
    
    nbr_loan_channel_cc_office = customer_data[customer_data["CHANNEL_TYPE"]=="Credit and cash offices"].shape[0]
    nbr_loan_channel_cntry_wide = customer_data[customer_data["CHANNEL_TYPE"].isin(["Country-wide"])].shape[0]
    nbr_loan_channel_stone_regional = customer_data[customer_data["CHANNEL_TYPE"].isin(["Stone","Regional / Local", "Car dealer", "Contact center"])].shape[0]
    
    nbr_loan_high_yeild = customer_data[customer_data["NAME_YIELD_GROUP"]=="middle"].shape[0]
    nbr_loan_meddle_yeild = customer_data[customer_data["NAME_YIELD_GROUP"]=="high"].shape[0]
    nbr_loan_low_yeild = customer_data[customer_data["NAME_YIELD_GROUP"].isin(["low_normal", "low_action"])].shape[0]
    
    nbr_loan_POS_with_interset = customer_data[customer_data["PRODUCT_COMBINATION"].isin(["POS household with interest", "POS mobile with interest", "POS industry with interest", "POS other with interest"])].shape[0]
    nbr_loan_POS_without_interset = customer_data[customer_data["PRODUCT_COMBINATION"].isin(["POS household without interest", "POS mobile without interest", "POS industry without interest", "POS others without interest"])].shape[0]
    
    return total_loans_applied,total_loan_approved, total_loan_rejected, nbr_cash_loans, nbr_consumer_loans, nbr_revolving_loans, nbr_live_loans, nbr_closed_loans,\
           nbr_loan_with_downpayment,avg_percentage_downpayment,nbr_loan_rejection_HC, nbr_loan_rejection_LIMIT, nbr_loan_rejection_SCO,\
           nbr_loan_cross_sell, nbr_loan_walkin,nbr_loan_channel_cc_office,nbr_loan_channel_cntry_wide,nbr_loan_channel_stone_regional,\
           nbr_loan_high_yeild, nbr_loan_meddle_yeild, nbr_loan_low_yeild,nbr_loan_POS_with_interset,nbr_loan_POS_without_interset
    
```

```python
previous_application_group_categorical = previous_application.groupby("SK_ID_CURR", sort=False).apply(lambda previous_loan: get_grouped_data(previous_loan))
```

```python
previous_application_group_categorical =previous_application_group_categorical.reset_index()
```
