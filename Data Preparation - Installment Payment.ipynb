```python
def missing(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing(installment_payment)
```

```python
installment_payment_grouped_1  = installment_payment.groupby(["SK_ID_PREV","NUM_INSTALMENT_NUMBER"],as_index=False)['SK_ID_CURR',
       'NUM_INSTALMENT_VERSION', 'DAYS_INSTALMENT','DAYS_ENTRY_PAYMENT',
       'AMT_INSTALMENT'].max()
```

```python
installment_payment_grouped_2  = installment_payment.groupby(["SK_ID_PREV","NUM_INSTALMENT_NUMBER"],as_index=False)['AMT_PAYMENT'].sum()
```

```python
installment_payment_grouped_2[installment_payment_grouped_2["SK_ID_PREV"]== 1944356].sort_values("NUM_INSTALMENT_NUMBER")
```

```python
installment_payment_grouped = pd.merge(installment_payment_grouped_1, installment_payment_grouped_2, left_on= ["SK_ID_PREV","NUM_INSTALMENT_NUMBER"], right_on = ["SK_ID_PREV","NUM_INSTALMENT_NUMBER"])
```

```python
installment_payment_grouped.shape
```

```python
installment_payment_grouped["PAYMENT_RATIO"] = installment_payment_grouped.apply(lambda x: round(x["AMT_PAYMENT"]/x["AMT_INSTALMENT"],2), axis =1)
```

```python
installment_payment_grouped[installment_payment_grouped["PAYMENT_RATIO"] != 1]
```

```python
def get_payment_time(row):
    if row["DAYS_INSTALMENT"] >= row["DAYS_ENTRY_PAYMENT"]:
        return 1
    else:
        return 0
installment_payment_grouped["PAYMENT_TIME"] = installment_payment_grouped.apply(lambda x:get_payment_time(x), axis=1)
```

```python
def get_attributes(installment_details):
    cnt_on_time_payment = installment_details[installment_details["PAYMENT_TIME"]== 1].shape[0]
    cnt_on_delay_payment = installment_details[installment_details["PAYMENT_TIME"]== 0].shape[0]
    cnt_partial_payment = installment_details[installment_details["PAYMENT_RATIO"] < 1].shape[0]
    cnt_full_payment = installment_details[installment_details["PAYMENT_RATIO"] >= 1].shape[0]
    avg_installment_amount = installment_details["AMT_INSTALMENT"].mean()
    
    return max(installment_details["SK_ID_CURR"]),cnt_on_time_payment, cnt_on_delay_payment, cnt_partial_payment, cnt_full_payment, avg_installment_amount
    
    

installment_payment_overall = installment_payment_grouped.groupby("SK_ID_PREV", sort=False).apply(
    lambda installment: get_attributes(installment)
)
```

```python
installment_payment_overall = installment_payment_overall.reset_index()
```

```python
installment_payment_overall.head()
```

```python
installment_payment_overall[["SK_ID_CURR",'cnt_on_time_payment', 'cnt_on_delay_payment', 'cnt_partial_payment', 'cnt_full_payment', 'avg_installment_amount']] = pd.DataFrame(installment_payment_overall[0].tolist(), index=installment_payment_overall.index)  
```
